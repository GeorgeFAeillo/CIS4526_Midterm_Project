```python
#Parsing Train File Into CSV  
import csv
import nltk
from nltk.translate.bleu_score import sentence_bleu
from nltk.translate.nist_score import sentence_nist
from nltk.translate.meteor_score import single_meteor_score
from nltk.translate.bleu_score import SmoothingFunction
import pyter
from Levenshtein import distance as lev
smoothing_func = SmoothingFunction().method4 
with open('midterm_train.csv', 'w', newline='') as csv_file:
    fieldnames = ['bleu_1','bleu_2', 'bleu_3', 'bleu_4', 'nist_1','nist_2','nist_3','nist_4', 'nist_5','meteor', 'ter', 'lev','ground_truth']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()
    with open (r"C:\Users\gfaei\Downloads\MLMidTerm-main\MLMidTerm-main\train_with_label.txt",encoding='utf-8') as read_file:
        for line in read_file.readlines():
            line = line.split('\t')
            new_line_one = line[1]
            new_line_two = line[2]
            new_line_one = line[1].replace("’","'")
            new_line_one = new_line_one.replace(" 's ","'s ")
            new_line_one = new_line_one.replace(" ' s ","'s ")
            new_line_two = line[2].replace("’","'")
            new_line_two = new_line_two.replace(" 's ","'s ")
            new_line_two = new_line_two.replace(" ' s ","'s ")
            new_line_one_words = nltk.word_tokenize(new_line_one)
            if "." in new_line_one_words:
                new_line_one_words.remove(".")
            if "," in new_line_one_words:
                new_line_one_words.remove(",")
            new_line_two_words = nltk.word_tokenize(new_line_two)
            if "." in new_line_two_words:
                new_line_two_words.remove(".")
            if "," in new_line_two_words:
                new_line_two_words.remove(",")
            bleu_score_1 = sentence_bleu([new_line_one_words], new_line_two_words, smoothing_function=smoothing_func, weights = (1,0,0,0))
            bleu_score_2 = sentence_bleu([new_line_one_words], new_line_two_words, smoothing_function=smoothing_func, weights = (1./2.,1./2.))
            bleu_score_3 = sentence_bleu([new_line_one_words], new_line_two_words, smoothing_function=smoothing_func,weights = (1./3.,1./3.,1./3.))
            bleu_score_4 = sentence_bleu([new_line_one_words], new_line_two_words,smoothing_function=smoothing_func)
            nist_1 = sentence_nist(new_line_one_words,new_line_two_words,n=1)
            nist_2 = sentence_nist(new_line_one_words,new_line_two_words,n=2)
            nist_3 = sentence_nist(new_line_one_words,new_line_two_words,n=3)
            nist_4 = sentence_nist(new_line_one_words,new_line_two_words,n=4)
            nist_5 = sentence_nist(new_line_one_words,new_line_two_words,n=5)
            meteor = single_meteor_score(new_line_one_words,new_line_two_words)
            ter = pyter.ter(new_line_one_words,new_line_two_words)
            distance = lev(new_line_one,new_line_two)
            ground_truth = ""
            for value in line[3]:
                if value =='0' or value =='1':
                    ground_truth = value
            writer.writerow({"bleu_1":bleu_score_1,"bleu_2": bleu_score_2, "bleu_3": bleu_score_3, "bleu_4": bleu_score_4, "nist_1": nist_1, "nist_2": nist_2, "nist_3": nist_3, "nist_4": nist_4, "nist_5":nist_5,"meteor":meteor, "ter":ter, "lev":distance, "ground_truth": ground_truth})
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier
import pandas as pd
train_data_frame = pd.read_csv('midterm_train.csv')
#Since Ones dominate training data, add Zero's to make the training data have a similar distribution to dev and test
zeroes = train_data_frame[train_data_frame["ground_truth"]==0]
for i in range(3):
    train_data_frame= train_data_frame.append(zeroes,ignore_index=True)
dev_data_frame = pd.read_csv('midterm_dev.csv')
scaler = MinMaxScaler()
y_train_val = train_data_frame["ground_truth"].values
x_train_val = scaler.fit_transform(train_data_frame.drop('ground_truth', axis=1).values)
y_dev_val = dev_data_frame["ground_truth"].values
x_dev_val = scaler.transform(dev_data_frame.drop('ground_truth', axis=1).values)
clf_lr = LogisticRegression(penalty='l1',solver='saga',C=0.5)
clf_lr.fit(x_train_val, y_train_val)
clf = SVC(kernel='linear')
clf.fit(x_train_val,y_train_val)
clf_r = SVC()
clf_r.fit(x_train_val,y_train_val)
abc = VotingClassifier(estimators=[('lr', clf_lr), ('svm_l', clf), ('svm', clf_r)], voting='hard')
abc.fit(x_train_val,y_train_val)
y_pred = abc.predict(x_dev_val)
train_pred = abc.predict(x_train_val)
acc_train = accuracy_score(y_train_val,train_pred)
print("ACC TRAIN VOTING")
print(acc_train)
acc = accuracy_score(y_dev_val, y_pred)
print("ACC DEV VOTING")
print(acc)
```
